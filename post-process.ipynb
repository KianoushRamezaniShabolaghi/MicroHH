 a code to print all of the .nc files

```python
# ────────────────────── list_nc_files.py ──────────────────────
import glob
from pathlib import Path

# Find ALL .nc files in the current directory
nc_files = sorted(glob.glob("*.nc"))

if not nc_files:
    print("No .nc files found in this directory!")
else:
    print(f"Found {len(nc_files)} NetCDF (.nc) files:\n")
    print("-" * 70)
    
    for f in nc_files:
        path = Path(f)
        size_mb = path.stat().st_size / (1024**2)  # size in MB
        
        # Try to extract time from filename like drycblles.0036000.nc → 10.00 h
        time_str = None
        if "drycblles" in f or any(x in f for x in ["0000000", "001800", "003600"]):
            # Look for 7-digit number before .nc
            import re
            match = re.search(r'(\d{7})\.nc$', f)
            if match:
                t_sec = int(match.group(1))
                t_hour = t_sec / 3600.0
                time_str = f" → {t_sec:5d} s = {t_hour:5.2f} h"
        
        size_str = f"{size_mb:6.1f} MB"
        print(f"{f:45}  {size_str}  {time_str or ''}")

    print("-" * 70)
    print("Done! Ready for plotting with xarray")
# ─────────────────────────────────────────────────────────────────
```

###  create files like drycblles.0001800.nc and all of them

```python
# create_drycblles_nc_files.py
# Works perfectly for your 10-hour run with output every 1800 s

import numpy as np
from netCDF4 import Dataset
import glob
import os

# === Grid (your domain: 3200 m, 64×64×64 grid → 50 m spacing) ===
nx = ny = nz = 64
x = np.linspace(0, 3200, nx, endpoint=False)
y = np.linspace(0, 3200, ny, endpoint=False)
z = np.linspace(0, 3200, nz, endpoint=False)   # will be replaced with real z later if needed

# === Automatically find ALL available time steps from s001.* files ===
s001_files = sorted(glob.glob("s001.[0-9][0-9][0-9][0-9][0-9][0-9][0-9]"))
if not s001_files:
    raise FileNotFoundError("No s001.XXXXXXX files found! Are you in the right folder?")

times = []
for f in s001_files:
    t_str = f.split(".")[-1]
    if len(t_str) == 7 and t_str.isdigit():
        times.append(int(t_str))

times = sorted(times)
print(f"Found {len(times)} time steps: from {times[0]} s to {times[-1]} s "
      f"({times[-1]/3600:.1f} h)")

# === Loop over all times and create one .nc file per time step ===
for t in times:
    tstr = f"{t:07d}"
    print(f"Creating drycblles.{tstr}.nc ...")

    # Load the five fields
    u  = np.fromfile(f"u.{tstr}",     dtype=np.float64).reshape((nz, ny, nx))
    v  = np.fromfile(f"v.{tstr}",     dtype=np.float64).reshape((nz, ny, nx))
    w  = np.fromfile(f"w.{tstr}",     dtype=np.float64).reshape((nz, ny, nx))
    th = np.fromfile(f"th.{tstr}",    dtype=np.float64).reshape((nz, ny, nx))
    s001 = np.fromfile(f"s001.{tstr}", dtype=np.float64).reshape((nz, ny, nx))

    # Create NetCDF file
    nc = Dataset(f"drycblles.{tstr}.nc", "w", format="NETCDF4")

    nc.createDimension("x", nx)
    nc.createDimension("y", ny)
    nc.createDimension("z", nz)
    nc.createDimension("time", None)

    # Coordinates
    nc.createVariable("x", "f8", ("x",))[:] = x
    nc.createVariable("y", "f8", ("y",))[:] = y
    nc.createVariable("z", "f8", ("z",))[:] = z
    nc.createVariable("time", "f8", ("time",))[:] = t
    nc.variables["time"].units = "seconds since simulation start"

    # Variables
    for name, data in zip(["u", "v", "w", "th", "s001"], [u, v, w, th, s001]):
        var = nc.createVariable(name, "f8", ("time", "z", "y", "x"), zlib=True, complevel=4)
        var[0, :, :, :] = data
        var.units = "m/s" if name in ["u","v","w"] else "K" if name == "th" else "kg/m³ (or arbitrary)"

    nc.close()
    print(f"  → drycblles.{tstr}.nc created ({t/3600:.2f} h)")

print("\nAll done! Created", len(times), "files from 0 to 10 hours.")
print("You can now run the beautiful 4-panel plot!")
```

```python
# After conversion, run this (the same plot code from before)
import glob, xarray as xr, matplotlib.pyplot as plt, numpy as np

files = sorted(glob.glob("drycblles.*.nc"))
ds = xr.open_mfdataset(files, combine='by_coords', decode_times=False)

c = ds['s001']
z = ds['z']
time_h = ds['time'] / 3600

total = c.sum(dim=['x','y','z'])
z_eff = (c * z).sum(dim=['x','y','z']) / total
sigma_y = np.sqrt(((ds.y - (c*ds.y).sum(dim=['x','y','z'])/total)**2 * c).sum(dim=['x','y','z']) / total)
sigma_z = np.sqrt(((z - z_eff)**2 * c).sum(dim=['x','y','z']) / total)
glc_max = c.isel(z=0).max(dim=['x','y'])

fig, axs = plt.subplots(2,2,figsize=(13,10))
axs[0,0].plot(time_h, z_eff, 'o-', lw=3, color='red'); axs[0,0].set_ylabel('z_eff [m]')
axs[0,1].plot(time_h, sigma_y, '^-', lw=3, label='σ_y'); axs[0,1].plot(time_h, sigma_z, 's-', lw=3, label='σ_z'); axs[0,1].legend()
axs[1,0].semilogy(time_h, glc_max, 'D-', color='purple'); axs[1,0].set_ylabel('Max GLC')
c.isel(time=-1,z=0).plot.contourf(ax=axs[1,1], levels=30, cmap='viridis')
plt.tight_layout(); plt.show()
```

```python
import glob
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# Only select files that match the pattern: drycblles.XXXXXXX.nc (7 digits)
files = sorted([f for f in glob.glob("drycblles.*.nc") 
                if f.count('.') == 2 and len(f.split('.')[-2]) == 7 and f.split('.')[-2].isdigit()])

print(f"Found {len(files)} valid 3D snapshot files:")
for f in files: print("  ", f)

heights = []
sig_y   = []
sig_z   = []
times_h = []

for f in files:
    ds = xr.open_dataset(f, decode_times=False)
    
    c = ds['s001'].isel(time=0)                    # shape (z,y,x)
    t_h = float(ds.time.values[0]) / 3600.0
    
    c_total = c.sum(dim=['x', 'y', 'z'])
    
    zeff       = (c * ds.z).sum(dim=['x','y','z']) / c_total
    y_centroid = (c * ds.y).sum(dim=['x','y','z']) / c_total
    
    cy = ds.y - y_centroid
    cz = ds.z - zeff
    
    sigma_y = np.sqrt( (c * cy**2).sum(dim=['x','y','z']) / c_total )
    sigma_z = np.sqrt( (c * cz**2).sum(dim=['x','y','z']) / c_total )
    
    heights.append(float(zeff))
    sig_y.append(float(sigma_y))
    sig_z.append(float(sigma_z))
    times_h.append(t_h)
    
    print(f"{f[-15:]:15} → t = {t_h:5.2f} h → z_eff = {zeff.values:6.1f} m, "
          f"σ_y = {sigma_y.values:5.0f} m, σ_z = {sigma_z.values:5.0f} m")
    
    ds.close()

# =================================================================
# Beautiful plot
plt.figure(figsize=(10, 6))
plt.plot(times_h, heights, 'o-', color='tab:red', lw=3, ms=8, label=r'$z_{\mathrm{eff}}$ (LES)')
plt.plot(times_h, sig_z,   's--', color='tab:blue', lw=2.5, ms=7, label=r'$\sigma_z$ (LES)')
plt.plot(times_h, sig_y,   '^-',  color='tab:green', lw=2, ms=6, label=r'$\sigma_y$ (LES)')

plt.xlabel('Time [h]', fontsize=12)
plt.ylabel('Height / Spread [m]', fontsize=12)
plt.title('Plume Rise and Spreading in Dry Convective Boundary Layer', fontsize=14, pad=15)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import glob
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# File selection
files = sorted([f for f in glob.glob("drycblles.*.nc")
                if f.count('.') == 2 and len(f.split('.')[-2]) == 7 and f.split('.')[-2].isdigit()])

times_h = []
max_glc = []
mean_glc = []
z0_glc = []
selected_files = []

print("Loading data – only 0.5 ≤ t ≤ 10.0 h ...\n")

for f in files:
    ds = xr.open_dataset(f, decode_times=False)
    t_h = float(ds.time[0].values) / 3600.0
    
    if t_h < 0.5 or t_h > 10.0:
        ds.close()
        continue
    
    c_ground = ds['s001'].isel(time=0, z=0)
    z_ground = float(ds.z[0].values)
    
    max_val = float(c_ground.max())
    mean_val = float(c_ground.mean())
    
    max_glc.append(max_val)
    mean_glc.append(mean_val)
    times_h.append(t_h)
    z0_glc.append(z_ground)
    selected_files.append(f)
    
    print(f"{f[-15:]:15} → t = {t_h:5.2f} h → max = {max_val:8.2e}")
    ds.close()

if len(times_h) == 0:
    raise ValueError("No files found between 0.5 h and 10 h!")

# ================================================================
# Smart y-axis limits (fixed!)
# ================================================================
all_values = np.array(max_glc + mean_glc)          # combine both curves
positive_values = all_values[all_values > 0]        # ignore zeros
ymin_data = positive_values.min()
ymax = np.max(max_glc)

# Set nice limits with a little breathing room
ymin_plot = ymin_data / 10          # one decade below smallest value
ymax_plot = ymax * 2                # extra space above peak

fig, ax = plt.subplots(figsize=(11, 6))

ax.semilogy(times_h, max_glc, 'o-', color='#d32f2f', lw=3.2, ms=8, 
            markerfacecolor='white', markeredgewidth=2,
            label=f'Maximum GLC @ z ≈ {z0_glc[0]:.1f} m')

ax.semilogy(times_h, mean_glc, 's--', color='#f57c00', lw=2.8, ms=7,
            label='Domain-averaged GLC')

ax.set_xlabel('Time [h]', fontsize=13)
ax.set_ylabel('Ground-Level Concentration [kg m⁻³]', fontsize=13)
ax.set_title('Evolution of Near-Surface Plume Concentration\nDry Convective Boundary Layer LES — 0.5 to 10 h', 
             fontsize=15, pad=20)

ax.set_xlim(0.5, 10.0)
ax.set_ylim(ymin_plot, ymax_plot)

ax.grid(True, which="both", ls=":", alpha=0.6, lw=0.8)
ax.legend(fontsize=12, frameon=True, fancybox=True, edgecolor='black')

plt.tight_layout()
plt.show()

# ================================================================
# Final-time contour plot (improved color scaling)
# ================================================================
ds_final = xr.open_dataset(selected_files[-1], decode_times=False)
c_final = ds_final['s001'].isel(time=0, z=0)
t_final_h = float(ds_final.time[0].values) / 3600.0

levels = np.logspace(np.log10(c_final.quantile(0.01)), np.log10(c_final.max()), 40)

plt.figure(figsize=(9, 7.5))
cf = c_final.plot.contourf(levels=levels, cmap='inferno', extend='max')
plt.title(f"Ground-Level Concentration at t = {t_final_h:.2f} h\n(z ≈ {ds_final.z[0].values:.1f} m)", 
          fontsize=15, pad=15)
plt.xlabel('x [m]', fontsize=12)
plt.ylabel('y [m]', fontsize=12)

cb = plt.colorbar(cf, pad=0.02, shrink=0.8)
cb.set_label('Concentration [kg m⁻³]', rotation=90, labelpad=12, fontsize=12)

plt.tight_layout()
plt.show()

ds_final.close()

# Optional: save both figures automatically
fig.savefig("GLC_time_series_0.5_to_10h.pdf", dpi=300, bbox_inches='tight')
plt.savefig("GLC_time_series_0.5_to_10h.png", dpi=300, bbox_inches='tight')
print("Figures saved as GLC_time_series_0.5_to_10h.pdf and .png")

plt.gcf().savefig("GLC_contour_t10h.pdf", dpi=300, bbox_inches='tight')
plt.gcf().savefig("GLC_contour_t10h.png", dpi=300, bbox_inches='tight')
print("Contour plots saved as GLC_contour_t10h.pdf and .png")
```
